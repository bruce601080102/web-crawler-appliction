```python
import requests
import time
import json
from bs4 import BeautifulSoup

URL='https://www.female007.com.tw/'
URL_html =requests.get(URL)
URL_html.encoding='utf-8' 

All_Url=[]
articles_top=[]
articles_bottom=[]
name_title=[]  
area_title=[]
munber_list=[]
set_position_top=[]
set_position_bottom=[]
correct_position_top=[]
correct_position_bottom=[]

soup_html = BeautifulSoup(URL_html.text)
soup_html = soup_html.find_all('div','f_foot')
for soup_html in soup_html:
    soup_html=soup_html.find_all('a')
    
for soup_html_href in soup_html:
    All_Url.append(soup_html_href.get('href'))
    
Menu_Url=All_Url[0:9]
Area_Url=All_Url[-15:]    
Area_Url=Menu_Url+Area_Url   

for soup_html_title in soup_html:
    soup_html_title=soup_html_title.get('title')
    name_title.append(soup_html_title)
    
name_title=name_title[0:9]

for soup_html in soup_html:
    soup_html=soup_html.text.strip()
    area_title.append(soup_html)
    
area_title=area_title[-15:]
title=name_title+area_title

for number in range(1,25):
    munber_list.append(number)
    
dict_name=dict(zip(munber_list,title))

time=0

for a in Area_Url:
    time+=1
    print('')
    print('')
    print('============================================='+str(time)+'=======================================')
    print('\033[1;31m============================================='+dict_name.get(time)+'=======================================\033[0m')
    print('--------------'+str(a)+'--------------')
    print('**********************************************\033[4;34m上半部分\033[0m***************************************************')
    URL_html=URL+str(a)
    print(URL_html)
    req =requests.get(URL_html)
    req.encoding='utf-8' 
    soup = BeautifulSoup(req.text)
    #print(soup)
    teltop = soup.find_all('div','teltop')
    
    for teltop in teltop:
        teltop=teltop.find('div','tel_inside')
        print(teltop.text.strip())
        teltop=teltop.text.strip()
        articles_top.append(teltop)#上半部分
    print('**********************************************\033[4;31m下半部分\033[0m!***************************************************')  
    f_foot = soup.find_all('div','f_foot')
    for f_foot in f_foot:
        f_foot=f_foot.find('div','tel_inside')
        print(f_foot.text.strip())
        f_foot=f_foot.text.strip()   
        articles_bottom.append(f_foot)
```

```python
print('上半部各頁電話出現的次數:'+str(articles_top.count(teltop))+'/24')
print('下半部各頁電話出現的次數:'+str(articles_bottom.count(f_foot))+'/24')
print('='*40)
if articles_top.count(teltop)!=24:
    
    for correct_top in correct_position_top:
        set_position_top.remove(correct_top) 

    for wrong_top in set_position_top:
        print('錯誤區域(上)-->'+dict_name.get(wrong_top))
        
    if  articles_bottom.count(f_foot)!=24: 
        
        for correct_bottom in correct_position_bottom:
            set_position_bottom.remove(correct_bottom) 

        for wrong_bottom in set_position_bottom:
            print('錯誤區域(下)-->'+dict_name.get(wrong_bottom))    
        
elif articles_bottom.count(f_foot)!=24:
    
    for correct_bottom in correct_position_bottom:
        set_position_bottom.remove(correct_bottom) 

    for wrong_bottom in set_position_bottom:
        print('錯誤區域(下)-->'+dict_name.get(wrong_bottom)) 
        
    if articles_top.count(teltop)!=24:

        for correct_top in correct_position_top:
            set_position_top.remove(correct_top) 

        for wrong_top in set_position_top:
            print('錯誤區域(上)-->'+dict_name.get(wrong_top))     
    
    
    
        
else :
    print('今天的電話比對完全正確')
```
